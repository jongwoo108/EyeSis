```python
from ultralytics import YOLO
model = YOLO("yolo11n.pt")
```

```python
# 2. 이미지 경로 지정
image_path = "test.jpg"
```

```python
# 3. 추론 실행
results = model(image_path)   # 리스트 형태로 결과가 들어옴
```

```python
# 4. 결과 시각화 + 파일로 저장
for i, r in enumerate(results):
    # 결과를 이미지로 저장 (runs/detect/predict/ 안에 저장됨)
    r.save()  

print("완료! runs/detect/predict 폴더에 결과 이미지가 생겼을 거야.")
```
